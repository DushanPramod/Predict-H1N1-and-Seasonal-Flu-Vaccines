```python
from pathlib import Path

import numpy as np
import pandas as pd

pd.set_option("display.max_columns", 100)
```

```python
DATA_PATH = Path.cwd()
```

```python
features_df = pd.read_csv(DATA_PATH / "training_set_features.csv", index_col="respondent_id")
labels_df = pd.read_csv(DATA_PATH / "training_set_labels.csv", index_col="respondent_id")
```

```python
print("features_df.shape", features_df.shape)
features_df.head()
```

```python
features_df.dtypes
```

```python
print("labels_df.shape", labels_df.shape)
labels_df.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

from sklearn.linear_model import LogisticRegression
from sklearn.multioutput import MultiOutputClassifier

from sklearn.pipeline import Pipeline

from sklearn.model_selection import train_test_split

from sklearn.metrics import roc_curve, roc_auc_score

from sklearn.preprocessing import OneHotEncoder

RANDOM_SEED = 6    # Set a random seed for reproducibility!
```

```python
features_df.dtypes != "object"
```

```python
numeric_cols = features_df.columns[features_df.dtypes != "object"].values
print(numeric_cols)
```

```python
categorical_cols = features_df.columns[features_df.dtypes == "object"].values
print(categorical_cols)
```

```python
# chain preprocessing into a Pipeline object
# each step is a tuple of (name you chose, sklearn transformer)
numeric_preprocessing_steps = Pipeline([
    ('standard_scaler', StandardScaler()),
    ('simple_imputer', SimpleImputer(strategy='median'))
])

categorical_transformer = OneHotEncoder(handle_unknown="error")

# create the preprocessor stage of final pipeline
# each entry in the transformer list is a tuple of
# (name you choose, sklearn transformer, list of columns)
preprocessor = ColumnTransformer(
    transformers = [
        ("categorical", categorical_transformer, categorical_cols),
        ("numeric", numeric_preprocessing_steps, numeric_cols),
    ],
    remainder = "drop"
)
```

```python
estimators = MultiOutputClassifier(
    estimator=LogisticRegression(penalty="l2", C=1, max_iter=1000)
)
```

```python
full_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("estimators", estimators),
])
```

```python
full_pipeline
```

```python
X_train, X_eval, y_train, y_eval = train_test_split(
    features_df,
    labels_df,
    test_size=0.33,
    shuffle=True,
    stratify=labels_df,
    random_state=RANDOM_SEED
)
```

```python
%%time

# Train model
full_pipeline.fit(X_train, y_train)


# Predict on evaluation set
# This competition wants probabilities, not labels
preds = full_pipeline.predict_proba(X_eval)
preds
```

```python
print("test_probas[0].shape", preds[0].shape)
print("test_probas[1].shape", preds[1].shape)
```

```python
y_preds = pd.DataFrame(
    {
        "h1n1_vaccine": preds[0][:, 1],
        "seasonal_vaccine": preds[1][:, 1],
    },
    index = y_eval.index
)
print("y_preds.shape:", y_preds.shape)
y_preds.head()
```
